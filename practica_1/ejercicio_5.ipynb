# Ejercicio 5: Tabla de multiplicar

**Enunciado:** generar la tabla de multiplicar de un número entero del 1 al 12.

**Lógica:** se define un número entero y se utiliza un ciclo `for` para repetir el proceso desde el 1 hasta el 12. En cada repetición, el número se multiplica por el valor actual del ciclo y se muestra el resultado.

```python
def tabla_multiplicar(numero):
    for i in range(1, 13):
        resultado = numero * i
        print(f"{numero} x {i} = {resultado}")


numero = 9

print(f"Tabla de multiplicar del {numero}")
tabla_multiplicar(numero)
```
