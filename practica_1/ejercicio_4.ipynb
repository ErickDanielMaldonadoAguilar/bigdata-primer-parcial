# Ejercicio 4: Número par o impar

**Enunciado:** determinar si un número entero es par o impar.

**Lógica:** se utiliza el operador módulo `%`, el cual permite obtener el residuo de una división. Si un número dividido entre 2 deja residuo 0, entonces es par. Si deja residuo diferente de 0, entonces es impar.

```python
def verificar_par_impar(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"


numero = 15
resultado = verificar_par_impar(numero)

print(f"Número: {numero}")
print(f"Resultado: El número es {resultado}")
```
